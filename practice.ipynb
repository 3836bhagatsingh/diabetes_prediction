```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.datasets import load_diabetes
```

```python
diabetes = load_diabetes()
diabetes
```

```python
diabetes.data
```

```python
diabetes.target
```

```python
df = pd.DataFrame(data=diabetes.data,columns=diabetes.feature_names)
```

```python
df
```

```python
df['target'] = diabetes.target
```

```python
df
```

```python
x = df.drop(columns=['target'])
x
```

```python
y = df['target']
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) 
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
x_train_scaled.shape,x_test_scaled.shape
```

```python
ridge_model = Ridge()
lasso_model = Lasso()
```

```python
ridge_model.fit(x_train_scaled,y_train)
```

```python
lasso_model.fit(x_train_scaled,y_train)
```

```python
ridge_model_preiction = ridge_model.predict(x_test_scaled)
```

```python
ridge_model_preiction
```

```python
lasso_model_prediction = lasso_model.predict(x_test_scaled)
```

```python
lasso_model_prediction
```

```python
ridge_error = mean_squared_error(y_test,ridge_model_preiction)
lasso_error = mean_squared_error(y_test,lasso_model_prediction)
```

```python
ridge_error
```

```python
lasso_error
```

```python
from sklearn.metrics import r2_score,root_mean_squared_error
```

```python
r2_score(y_test,ridge_model_preiction)
```

```python
r2_score(y_test,lasso_model_prediction)
```
